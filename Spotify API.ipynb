```python
def getTrackAudioAnalyze_sections(id):
    meta = sp.track(id)
    audio_analyze = sp.audio_analysis(id)
    sections = audio_analyze['sections']
    return sections

def getTrackAudioAnalyze_segments(id):
    meta = sp.track(id)
    audio_analyze = sp.audio_analysis(id)
    segments = audio_analyze["segments"]
    return segments

def getTrackAudioAnalyze_beats(id):
    meta = sp.track(id)
    audio_analyze = sp.audio_analysis(id)
    beats = audio_analyze["beats"]
    return beats

def getTrackAudioAnalyze_bars(id):
    meta = sp.track(id)
    audio_analyze = sp.audio_analysis(id)
    bars = audio_analyze["bars"]
    return bars

def getPlaylistTrackIDs(user, playlist_id):
    ids = []
    playlist = sp.user_playlist(user, playlist_id)
    for item in playlist['tracks']['items']:
        track = item['track']
        ids.append(track['id'])
    return ids

ids = getPlaylistTrackIDs('dottyjunkie', '37i9dQZF1DZ06evO2JZXYI')
print(ids)
```

```python
def get_boundarues(df_track_section):
    df_plot = df_track_section['start'].tolist()
    temp = df_track_section['duration'].tolist()

    df_plot[0]=0.01
    df_plot.append(df_plot[-1] + temp[-1])
    return df_plot
```

```python
df_plot_01=get_boundarues(df_track_1_section)
df_plot_02=get_boundarues(df_track_2_section)
```

```python
def drawTrackSections(df_plot):
#   set fig size
    fig, ax = plt.subplots(figsize=(10, 2))
    fig.subplots_adjust(bottom=0.5)
    
#   draw colorbar
    cmap = mpl.cm.cool
    cb2 = mpl.colorbar.ColorbarBase(ax, 
                                    cmap=cmap,
                                    boundaries=df_plot,
                                    extend='neither',
                                    spacing='proportional',
                                    orientation='horizontal')
    cb2.set_label('virtual riot pray for riddim section in (sec)')
    return fig.show()
```

```python
  # get song info and audio analysis from song ids
def getTrackFeatures(id):
    meta = sp.track(id)
    features = sp.audio_features(id)

    # Meta
    name = meta['name']
    album = meta['album']['name']
    artist = meta['album']['artists'][0]['name']
    release_date = meta['album']['release_date']
    length = meta['duration_ms']
    popularity = meta['popularity']

    # Features
    acousticness = features[0]['acousticness']
    danceability = features[0]['danceability']
    energy = features[0]['energy']
    instrumentalness = features[0]['instrumentalness']
    liveness = features[0]['liveness']
    loudness = features[0]['loudness']
    speechiness = features[0]['speechiness']
    tempo = features[0]['tempo']
    time_signature = features[0]['time_signature']

    track = [name, album, artist, release_date, length, popularity, danceability, acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, tempo, time_signature]
    return track
```
