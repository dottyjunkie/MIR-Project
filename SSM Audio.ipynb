```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import librosa
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
from pylab import savefig
```

```python
###########
y, sr = librosa.load("Crankdat_drop.wav")
# y, sr = librosa.load("pray_drop.wav")
```

```python
y_64=np.reshape(y, (64, 8820)).T
y_32=np.reshape(y, (32, 17640)).T
y_16=np.reshape(y, (16, 35280)).T
```

```python
y_64=pd.DataFrame(y_64)
y_32=pd.DataFrame(y_32)
y_16=pd.DataFrame(y_16)
```

```python
def heatMap(df):
    #Create Correlation df
    corr = df.corr()
    #Plot figsize
    fig, ax = plt.subplots(figsize=(30, 30))
    #Generate Color Map
    colormap = sns.diverging_palette(220, 10, as_cmap=True)
    #Generate Heat Map, allow annotations and place floats in map
    sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
    #Apply xticks
    plt.xticks(range(len(corr.columns)), corr.columns);
    #Apply yticks
    plt.yticks(range(len(corr.columns)), corr.columns)
    #show plot
    plt.show()
```

```python
#64 x 64
heatMap(y_64)
```
